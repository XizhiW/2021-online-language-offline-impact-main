```python
import pandas as pd
import sys
import time

global INF 
INF = 99999
class graph():

    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[sys.maxsize for column in range(vertices)]
                    for row in range(vertices)]
        self.distance = [[sys.maxsize for column in range(vertices)]
                        for row in range(vertices)]
        self.Next = [[-1 for column in range(vertices)]
                    for row in range(vertices)]
    
    def constructPath(self, u, v):
        
        if (self.Next[u][v] == -1):
            return {}
    
        path = [u]
        while (u != v):
            u = self.Next[u][v]
            path.append(u)
    
        return path

    def printPath(self, path):
        n = len(path)
        for i in range(n):
            #print(path[i], end=" ")
            numberOfPath[path[i]]+=1
        # print (path[n - 1])
        # numberOfPath[path[n - 1]]+=1
    
    def printSolution(self, dist, start, dest):
        print("Vertex \tDistance from Source")
        print(dest, "\t", dist[start][dest])
        print("Path: ")

    def floyd_all(self):
        for i in range(self.V):
            for j in range(self.V):
                self.distance[i][j] = self.graph[i][j]
    
                if (self.graph[i][j] == sys.maxsize):
                    self.Next[i][j] = -1
                else:
                    self.Next[i][j] = j
             
        for k in range(self.V):
            for i in range(self.V):
                for j in range(self.V):
                    
                    if (self.distance[i][k] == sys.maxsize or self.distance[k][j] == sys.maxsize):
                        continue
                    if (self.distance[i][j] > self.distance[i][k] + self.distance[k][j]):
                        self.distance[i][j] = self.distance[i][k] + self.distance[k][j]
                        self.Next[i][j] = self.Next[i][k]
        # i=1
        # j=20
        # path=self.constructPath(i, j)
        # self.printPath(path)
        for i in range(self.V):
            for j in range(self.V):
                if i==j:
                    continue
                path = self.constructPath(i, j)
                self.printPath(path)
        
 
df=pd.read_csv("sample_data/Project 4_Problem 1_InputData.csv")
matrix_length=int(df['NodeID'].tail(1))+1
g= graph(matrix_length)
numberOfPath = [0] * matrix_length
for count in range(matrix_length):
    g.graph[count][count]=0
for i in range(len(df)):
    g.graph[df['NodeID'][i]][df['ConnectedNodeID'][i]]=df['Distance'][i]
#print(g.graph[0])
# starting_node=int(input("Please input starting node: "))
# dest_node=int(input("Please input destination node: "))
# start_time = time.time()

g.floyd_all()
#print("--- %s seconds ---" % (time.time() - start_time))
#print("InputFile10")

N=20
res = sorted(range(len(numberOfPath)), key = lambda sub: numberOfPath[sub])[-N:][::-1]
print("Top 20 vertices from 20th to 1st: "+str(res))

# for i in range(len(res)):
#     print(numberOfPath[res[i]])
```
