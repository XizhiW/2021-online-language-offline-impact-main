```python
dataframe_path_530 = 'mapping_US/dataframe_mapping_0531.csv'
```

```python
dataframe_530 = pd.read_csv(dataframe_path_530)
```

```python
dataframe_530
```

```python
dataframe_530=dataframe_530.drop(['Unnamed: 0'],axis=1)
```

```python
dataframe_530
```

```python
from numpy.core.numeric import NaN
print(dataframe_530['Mapping'][1])
if(pd.isna(dataframe_530['Mapping'][1])):
  print(1)
```

```python
del_list=[]
num=0
for i in dataframe_530['Mapping']:
  if(pd.isna(i)):
    del_list.append(num)
  num+=1
```

```python
for drop_index in del_list:
  dataframe_530=dataframe_530.drop(labels=drop_index,axis=0)
```

```python
dataframe_530.reset_index(drop=True, inplace=True)
```

```python
pd.set_option('display.max_rows', None)
```

```python
dataframe_result=dataframe_530
```

```python
list_citya=dataframe_result['City'].tolist()
list_numa=dataframe_result['Number'].tolist()
list_lata=dataframe_result['Latitude'].tolist()
list_lona=dataframe_result['Longtitude'].tolist()
list_addressa=dataframe_result['Mapping'].tolist()
```

```python
for i in range(len(list_citya)):
  j=i+1
  while(j<len(list_citya)):
    if(list_lata[i]==list_lata[j]):
      if(list_lona[i]==list_lona[j]):
        list_numa[i]+=list_numa[j]
        del list_citya[j]
        del list_numa[j]
        del list_lata[j]
        del list_lona[j]
        del list_addressa[j]
        continue
    j+=1
```

```python
dataframe_result
```

```python
dict_result_a={"City" : list_citya,
   "Number" : list_numa,
   "lat" : list_lata,
   "lon" : list_lona,
   "address" : list_addressa}
from pandas.core.frame import DataFrame
dataframe_result_a1=DataFrame(dict_result_a)
dataframe_result_a1
```
