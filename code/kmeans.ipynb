```python
import numpy as np
import pandas as pd
import random

class kmeans(object):
    def __init__(self, k=2, tolerance=0.1, max_iter=30):
        self.k=k
        self.tolerance=tolerance
        self.max_iter=max_iter

    def fit(self,data):
        #print("len_data: ",len(data))
        self.centroids_ = {}
        value_holder = []
        random_index=random.sample(list(range(len(data))),self.k)
        for i in random_index:
            value_holder.append(data[i])
            #print(data[i])
        for j in range(self.k):
            self.centroids_[j]=value_holder[j]
        #print("centroids: ",self.centroids_)
        prev_loss=999
        for i in range(self.max_iter):
            #print("Iteration: ", i)
            self.groups_ = {}
            for i in range(self.k):
                self.groups_[i]=[]
            #print("centroids: ",self.centroids_)
            for points in data:
                distances = []
                for center in self.centroids_:
                    distances.append(np.sqrt(np.sum((points-self.centroids_[center])**2)))
                classification = distances.index(min(distances))
                self.groups_[classification].append(points)

            #print("grouping: ",self.groups_)
            # prev_centroids = dict(self.centroids_)
            
            #print("centroids: ",self.centroids_)
            for group in self.groups_:
                self.centroids_[group]=np.average(self.groups_[group],axis=0)
            #print("centroids: ",self.centroids_)
            convergent = True
            total_loss=0
            for i in range(self.k):
                for points in self.groups_[i]:
                  total_loss+=sum((points-self.centroids_[i])**2)
                loss=total_loss/len(data)
            if abs(prev_loss-loss)>self.tolerance:
                convergent = False
                prev_loss=loss
            
            if convergent or i==self.max_iter-1:
                print("centroids: ",self.centroids_)
                print("converged at iteration ",i)
                print("loss: ",loss)
                for i in range(self.k):
                    print("number of points in group "+str(i)+": "+str(len(self.groups_[i])))
                break


if __name__ == '__main__':
    # test_case=pd.read_csv("Project3_Test_Case.csv")
    # #print(test_case["X1"])
    # list2=[]
    # for i in range(len(test_case)):
    #     list1=[]
    #     list1.append(test_case["X1"][i])
    #     list1.append(test_case["X2"][i])
    #     list2.append(list1)
    # x1 = np.array(list2)
    # #print(x1)
    # kmeans1 = kmeans(k=2)
    # kmeans1.fit(x1)

    power_case=pd.read_csv("Project3_Power_Consumption.csv")
    list4=[]
    for i in range(len(power_case)):
        list3=[]
        list3.append(power_case["Global_active_power"][i])
        list3.append(power_case["Global_reactive_power"][i])
        list3.append(power_case["Voltage"][i])
        list3.append(power_case["Global_intensity"][i])
        list3.append(power_case["Sub_metering_1"][i])
        list3.append(power_case["Sub_metering_2"][i])
        list3.append(power_case["Sub_metering_3"][i])
        list4.append(list3)
    x2 = np.array(list4)
    #print(x2)
    # kmeans1 = kmeans(k=2)
    # kmeans1.fit(x2)

    #for i in range(3,21):
    for i in range(12,21):
        print("-------i="+str(i)+"-------")
        kmeans2=kmeans(k=i)
        kmeans2.fit(x2)

```
